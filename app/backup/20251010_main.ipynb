#### Matriz Z (setores z setores)

```python
# Zι é a demanda intermediária total pelo produto de cada setor (soma das linhas de 𝑍, se 𝑧𝑖𝑗​ for a venda do setor 𝑖 para o setor 𝑗).

# Selecionando apenas setores
Z = mip_energia.loc[:73, '1':'73']
Z = np.array(Z)

print("Matriz Z (Fluxo entre setores):\n")
print("\n".join(" ".join(f"{v:>10.2f}" for v in row) for row in Z[:5, :5]))
```

#### Vetor de Demanda Final (f)

```python
# f é a demanda final (famílias, governo, FBCF, estoques, exportações).

# Vetor de Demanda Final (f)
f = mip_energia["81"].values.reshape(-1, 1)

print("\nVetor de Demanda Final (f):\n")
print("\n".join(f"{v:>10.2f}" for v in f.ravel()[:10]))
```

#### Vetor de Produção Total (x)

```python
#
# Vetor de Produção Total 
#
#  x = Z1 + f

# Calculando o vetor de produção total (x)
x_intermediario = Z.sum(axis=1).reshape(-1, 1)  # Soma dos fluxos intermediários (por linha) coluna df_z["Total Produto"]

# Calculando o vetor de produção total (x)
x = x_intermediario + f

# Exibindo os resultados
print("Vetor de Produção Total (x):")
print("\n".join(f"{v:>10.2f}" for v in x.ravel()[:10]))
```

```python
# Identificando o setor com maior produção
setor_maior_producao = mip_energia.loc[mip_energia["74"].idxmax(), "Descrição"]

print("\nSetor com maior produção:", setor_maior_producao)
```

#### Matriz de Coeficientes Técnicos (A)

```python
#
# Matriz de Coeficientes Técnicos 
#
# A = Z × x^-1
# 

A = Z / x

# Garantir que todos os valores sejam float64
A = A.astype(float)

# Exibindo as primeiras linhas da Matriz A
print("\nMatriz de Coeficientes Técnicos (A):\n")
print("\n".join("".join(f"{v:>10.4f}" for v in row) for row in A[:5, :5]))
```

```python
# Verificando as dimensões
print(f"Dimensão da Matriz Z: {Z.shape}")
print(f"\nDimensão do Vetor f: {f.shape}")
print(f"\nDimensão do Vetor x: {x.shape}")
print(f"\nDimensão do Vetor A: {A.shape}")
```

#### Matriz Inversa de Leontief (L)

```python
# Criando a Matriz Identidade (I) com o mesmo tamanho de A
I = np.eye(A.shape[0])

# Calculando a Inversa de Leontief (L)
try:
    L = np.linalg.inv(I - A)
except np.linalg.LinAlgError:
    print("A matriz (I - A) é singular e não pode ser invertida.")
    L = None

# Exibindo as primeiras linhas da Matriz L
if L is not None:
    print("\nMatriz Inversa de Leontief (L):")
    print("\n".join("".join(f"{v:>10.4f}" for v in row) for row in L[:5, :5]))
```

#### Valor Adicionado (VA)

```python
n = Z.shape[0]

# Calcular Valor Adicionado (VA)

# Fórmula teórica:
#     VA_i = x_i - Σ_j z_ij
VA = x - Z.sum(axis=1) 
```

### Multiplicador de Emprego

```python
x = np.asarray(x).reshape(-1)

emp = np.asarray(mip_emp).reshape(-1)
```

```python
# Evita divisão por zero em setores com x = 0 (se existirem)
x_safe = np.where(x == 0, np.nan, x)

# Coeficiente direto de emprego: empregos por R$ de produto
e = emp / x_safe  # (n,)

# Empregos totais por 1 R$ de demanda final em cada setor (tipo I)
#    Resultado é um vetor (n,), onde a posição j = empregos por 1 R$ de demanda final no setor j
M_emp = e @ L  # (n,)

# Escalas úteis
M_emp_per_million = 1e6 * M_emp  # empregos por R$ 1 milhão de demanda final

# 4) Multiplicador-razão Tipo I (Total/Direto) por setor
#    Direto por 1 R$ no setor j é e_j; então a razão é:
mult_tipoI_total_sobre_direto = M_emp / e

# 5) DataFrame organizado (par com o df_income que você já montou)
df_multi_emp = pd.DataFrame({
    "ID": np.arange(1, n+1),
    "Setor": variables["Descricao"][:n].to_numpy(),
    "Empregos": emp,
    "X": x,
    "Coef_Emprego": e,                       # empregos por R$ de produto (direto)
    "Emp_por_R$": M_emp,                     # empregos por R$ de demanda final (total: direto + indireto)
    "Emp_por_R$Mi": M_emp_per_million,       # empregos por R$ 1.000.000 de demanda final
    "Mult_TipoI_Total_sobre_Direto": mult_tipoI_total_sobre_direto
})
```

```python
df_multi_emp.to_excel("data\\multiplicadores\\multiplicadores_emprego.xlsx", index=None)
```

### Multiplicador de Produção

```python
#
# Total Output Multiplier
#
# Mi(total) = ∑j Lij
#

# Supondo que L (Matriz Inversa de Leontief) já está carregada corretamente
print("Dimensão da Matriz Inversa de Leontief (L):", L.shape)

# Calculando o Multiplicador Total de Produção para cada setor
multiplicadores_totais = L.sum(axis=1)  # Soma dos elementos de cada linha

# Dataframe
df_multi_total = pd.DataFrame({
    "ID": np.arange(1, n+1),
    "Setor": variables["Descricao"][:n].to_numpy(),
    "Multiplicador Produção": multiplicadores_totais
})

df_multi_total.to_excel("data\\multiplicadores\\multiplicadores_producao.xlsx", index=None)
```

#### Multiplicador Tipo I

```python
# Extrair o vetor de Demanda Final para os setores de energia (linha 81)
h = mip_energia.iloc[:, 81].values  # vetor (n,)
```

```python
# 
# Type I Multipliers
#
# M(I) = L × h 
# 
# Captura os efeitos diretos e indiretos na produção de cada setor como resultado de uma mudança na demanda final.

# Multiplicando a matriz inversa de Leontief pelo vetor h
M_type_I = L @ h

# Dataframe
df_multi_type_i = pd.DataFrame({
    "ID": np.arange(1, n+1),
    "Setor": variables["Descricao"][:n].to_numpy(),
    "Tipo I": M_type_I
})

df_multi_type_i.to_excel("data\\multiplicadores\\multiplicadores_tipo_i.xlsx", index=None)
```

#### Multiplicador Tipo II

```python
# 
# Type II Multipliers
#
# M(II) = L × (h + VA) 
# 

# Inclui os efeitos diretos, indiretos e induzidos na produção
M_type_II = L @ (h + VA)

# Dataframe
df_multi_type_ii = pd.DataFrame({
    "ID": np.arange(1, n+1),
    "Setor": variables["Descricao"][:n].to_numpy(),
    "Tipo I": M_type_I,
    "Tipo II": M_type_II
})

df_multi_type_ii.to_excel("data\\multiplicadores\\multiplicadores_tipo_ii.xlsx", index=None)
```
